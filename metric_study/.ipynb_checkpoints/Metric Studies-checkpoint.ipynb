# Studying CMFD metrics

This notebook is dedicated to test a new CMFD metric on a toy samples.\
This new metric considerates more consistency among the detected objects.\
Considering that each object and its copies are labeled with a unique ID, the metric \
should compare if at least two connected componnets from the detected map overlap the reference,\
otherwise this sould be considerated as inconsistent to CMFD.

# Import Cell

```python
LIB_PATH = "../"
import sys
sys.path.append(LIB_PATH)

from forgery_lib import (F1, precision, F1_CTP, precision_CTP)
from glob import glob
import cv2
from prettytable import PrettyTable
import base64
from scipy import ndimage
from PIL import Image
from io import BytesIO
from IPython.display import HTML
import io
def image_base64(img):
    im = img.copy()
    im[im>0] = 255
    im = Image.fromarray(im)
    with BytesIO() as buffer:
        im.save(buffer, 'jpeg')
        return base64.b64encode(buffer.getvalue()).decode()

def image_formatter(im):
    return f'<img src="data:image/jpeg;base64,{image_base64(im)}">'
%pylab inline
```

```python
def get_connected_components(cc_map,region_id):
    regions = np.zeros_like(cc_map)
    regions[cc_map == region_id] = 1
    regions_ids,ids = ndimage.label(regions)
    
    return regions_ids,ids
```

# Metrics

```python
def real_TP_img(img,gt):
    gt_cc, gt_ids = get_connected_components(gt,1)
    
    real_TP = np.zeros_like(gt)
    TP = 1
    for gi in range(1,gt_ids+1):

        gt_map = np.zeros_like(gt_cc)
        gt_map[gt_cc == gi] = 1
        gt_map = gt_map.astype('uint8')
        gt_i_score = cv2.bitwise_and(gt_map,img).sum()
        TP *= gt_i_score
    
    if TP > 0:
        for gi in range(1,gt_ids+1):

            gt_map = np.zeros_like(gt_cc)
            gt_map[gt_cc == gi] = 1
            gt_map = gt_map.astype('uint8')
            real_TP += cv2.bitwise_and(gt_map,img)
    return real_TP
```
